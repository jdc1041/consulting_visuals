```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
f1 = pd.read_excel("customers.xlsx")
f2 = pd.read_excel("contracts.xlsx")
```

```python
f1 = f1.rename(columns={'id': 'customer_id'})
```

```python
f1.set_index('customer_id', inplace=True)
f2.set_index('customer_id', inplace=True)
```

```python
df = f2.join(f1, how='inner')
```

```python
from matplotlib.pyplot import pie, axis, show
%matplotlib inline

```

```python
fig=plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.axis('equal')
a = df.groupby('industry').industry.count()
b = df.industry.unique()
ax.pie(a, labels=a.keys(), autopct='%1.1f%%')
fig = plt.gcf()
fig.set_size_inches(13,13) 
plt.title('Industries With Contracts')
plt.legend(b,loc='upper left',borderpad=0.00001)
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
bbox_to_anchor=(0.5,0.90)
plt.show()
```

Industry Pie Chart

This Chart is meant to show the difference between industries that take out contracts. While it is fairly large, it shows the different areas of where the contracts are extended to, and can help define the largest ones such as Major Pharmaceuticals and Major Banks. The business can take this information and use it to see which areas to increase contracts in, or cut them out fully and work to increase contracts in the major industries.
